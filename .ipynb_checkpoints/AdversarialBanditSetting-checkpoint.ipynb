```python
import math
import random
import numpy as np
import matplotlib.pyplot as plt

# Probability distribution function
def distr(weights, gamma, numActions):
    total_weight = sum(weights)
    base_prob = [(1 - gamma) * w / total_weight for w in weights]
    uniform_prob = gamma / numActions
    return [p + uniform_prob for p in base_prob]

# Draw function
def draw(probabilityDistribution):
    return np.random.choice(len(probabilityDistribution), p=probabilityDistribution)

# EXP3 Algorithm Implementation
def exp3(numActions, reward, gamma, rewardMin=0, rewardMax=1):
    weights = [1.0] * numActions
    t = 0
    while True:
        probabilityDistribution = distr(weights, gamma, numActions)
        choice = draw(probabilityDistribution)
        theReward = reward(choice, t)
        scaledReward = (theReward - rewardMin) / (rewardMax - rewardMin)
        estimatedReward = scaledReward / probabilityDistribution[choice]
        weights[choice] *= math.exp(estimatedReward * gamma / numActions)
        yield choice, theReward, estimatedReward, weights
        t += 1

# Simulation Function
def run_simulation(numActions, numRounds, gamma):
    biases = [1.0 / k for k in range(2, 12)]
    rewardVector = [[1 if random.random() < bias else 0 for bias in biases] for _ in range(numRounds)]
    rewards = lambda choice, t: rewardVector[t][choice]
    
    bestAction = max(range(numActions), key=lambda action: sum(rewardVector[t][action] for t in range(numRounds)))
    regrets = []
    cumulativeReward = 0
    bestActionCumulativeReward = 0
    
    for t, (choice, reward, _, _) in enumerate(exp3(numActions, rewards, gamma)):
        cumulativeReward += reward
        bestActionCumulativeReward += rewardVector[t][bestAction]
        regret = bestActionCumulativeReward - cumulativeReward
        regrets.append(regret)
        if t >= numRounds - 1:
            break
    
    return regrets

# Main Simulation and Visualization
numActions = 10
numRounds = 10000
num_runs = 1000
gamma = 0.07  # As per your code

# Run multiple times and average regret
all_regrets = np.array([run_simulation(numActions, numRounds, gamma) for _ in range(num_runs)])
expected_regret = np.mean(all_regrets, axis=0)

# Compute theoretical regret bound as a function of t
t_values = np.arange(1, numRounds + 1)
# Adjust bound to reflect O(sqrt(T K ln K)) for stochastic setting
optimal_gamma = np.sqrt(numActions * np.log(numActions) / t_values)  # Dynamic gamma for each t
theoretical_bound = (np.e - 1) * optimal_gamma * t_values + (numActions * np.log(numActions)) / optimal_gamma
# Use a simpler stochastic bound approximation: O(sqrt(T K ln K))
K = numActions
theoretical_bound = (np.e - 1) * np.sqrt(2 * K * t_values * np.log(K))

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(t_values, expected_regret, label='Expected Regret (Averaged over 1000 runs)', color='b', linewidth=2)
plt.plot(t_values, theoretical_bound, label='Theoretical Bound', color='r', linestyle='--', linewidth=2)
plt.xlabel('Time Step (t)', fontsize=12)
plt.ylabel('Cumulative Regret', fontsize=12)
plt.title('EXP3 Algorithm: Expected Regret vs Theoretical Bound', fontsize=14)
plt.legend(fontsize=10)
plt.grid(True)
plt.tight_layout()

# Save the plot
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

def exp3_algorithm(n, k, eta, num_runs=1000):
    # Store regrets across all runs
    all_regrets = np.zeros((num_runs, n))
    
    for run in range(num_runs):
        # Initialize variables
        S_hat = np.zeros(k)  # S_hat_0i = 0 for all i
        cumulative_rewards = np.zeros(k)
        best_cumulative_reward = 0
        cumulative_reward = 0
        
        for t in range(1, n+1):
            # Calculate sampling distribution P_t
            exp_values = np.exp(eta * S_hat)
            P_t = exp_values / np.sum(exp_values)
            
            # Sample action A_t and observe reward X_t
            A_t = np.random.choice(k, p=P_t)
            X_t = np.random.binomial(1, 1/(A_t+2))  # Using decreasing probabilities for arms
            
            # Update S_hat
            indicator = np.zeros(k)
            indicator[A_t] = 1
            S_hat = S_hat + (1 - (indicator * (1 - X_t))) / P_t
            
            # Calculate regret
            cumulative_reward += X_t
            cumulative_rewards += (np.arange(k) == A_t) * X_t
            best_action = np.argmax(cumulative_rewards)
            best_cumulative_reward = cumulative_rewards[best_action]
            regret = best_cumulative_reward - cumulative_reward
            all_regrets[run, t-1] = regret
    
    # Average over all runs
    expected_regret = np.mean(all_regrets, axis=0)
    
    # Theoretical bound: O(sqrt(n k log k))
    t_values = np.arange(1, n+1)
    theoretical_bound = np.sqrt(t_values * k * np.log(k))
    
    return expected_regret, theoretical_bound

# Parameters
n = 1200  # number of rounds
k = 10    # number of arms
eta = np.sqrt(np.log(k)/(n*k))  # learning rate

# Run algorithm
expected_regret, theoretical_bound = exp3_algorithm(n, k, eta)

# Plot results
plt.figure(figsize=(10, 6))
plt.plot(range(1, n+1), expected_regret, label='Expected Regret (Averaged over 1000 runs)')
plt.plot(range(1, n+1), theoretical_bound, label='Theoretical Bound', linestyle='--')
plt.xlabel('Time Step (t)')
plt.ylabel('Cumulative Regret')
plt.title('EXP3 Algorithm: Expected Regret vs Theoretical Bound')
plt.legend()
plt.grid(True)
plt.xlim(0, n)
plt.ylim(0, max(theoretical_bound[-1], expected_regret[-1]) * 1.1)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def run_exp3(n, k, eta):
    """Run EXP3 for `n` rounds and return cumulative regret at each step."""
    # True expected rewards for each arm (unknown to algorithm)
    true_means = np.array([1.0 / (i + 2) for i in range(k)])
    best_arm_mean = np.max(true_means)
    
    # Initialize variables
    S_hat = np.zeros(k)  # S_hat_0i = 0 for all i
    cumulative_reward = 0
    regret = np.zeros(n)
    
    for t in range(n):
        # Calculate sampling distribution P_t
        exp_values = np.exp(eta * S_hat)
        P_t = exp_values / np.sum(exp_values)
        
        # Sample action A_t and observe reward X_t
        A_t = np.random.choice(k, p=P_t)
        X_t = np.random.binomial(1, true_means[A_t])
        
        # Update S_hat
        indicator = np.zeros(k)
        indicator[A_t] = 1
        S_hat = S_hat + (1 - (indicator * (1 - X_t)) / P_t)
        
        # Update regret
        cumulative_reward += X_t
        regret[t] = (t + 1) * best_arm_mean - cumulative_reward
    
    return regret

def compute_expected_regret(n, k, eta, num_runs=1000):
    """Compute expected regret by averaging over `num_runs` runs."""
    all_regrets = np.zeros((num_runs, n))
    for run in range(num_runs):
        all_regrets[run] = run_exp3(n, k, eta)
    return np.mean(all_regrets, axis=0)

# Parameters
n = 1200  # number of rounds
k = 10    # number of arms
eta = np.sqrt(np.log(k) / (n * k))  # learning rate

# Compute expected regret (average over 1000 runs)
expected_regret = compute_expected_regret(n, k, eta)

# Theoretical bound: O(sqrt(n k log k))
t_values = np.arange(1, n + 1)
theoretical_bound = 2 * np.sqrt(t_values * k * np.log(k))  # Adjusted constant for visibility

# Plot results
plt.figure(figsize=(10, 6))
plt.plot(t_values, expected_regret, label='Expected Regret (Averaged over 1000 runs)', color='blue')
plt.plot(t_values, theoretical_bound, label='Theoretical Bound', linestyle='--', color='red')
plt.xlabel('Time Step (t)')
plt.ylabel('Cumulative Regret')
plt.title('EXP3 Algorithm: Expected Regret vs Theoretical Bound')
plt.legend()
plt.grid(True)
plt.xlim(0, n)
plt.ylim(0, max(theoretical_bound[-1], expected_regret[-1]) * 1.1)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Probability distribution function with type hints
def distr(weights: np.ndarray, gamma: float, num_actions: int) -> np.ndarray:
    """Compute EXP3 probability distribution with exploration."""
    total_weight = np.sum(weights)
    base_prob = (1 - gamma) * weights / total_weight
    uniform_prob = gamma / num_actions
    return base_prob + uniform_prob

# Draw function using NumPy's random choice
def draw(prob_dist: np.ndarray) -> int:
    """Draw an action based on the probability distribution."""
    return np.random.choice(len(prob_dist), p=prob_dist)

# EXP3 Algorithm Implementation
def exp3(num_actions: int, reward_func, gamma: float, reward_min: float = 0.0, reward_max: float = 1.0):
    """Generator for the EXP3 algorithm."""
    weights = np.ones(num_actions, dtype=float)
    t = 0
    while True:
        prob_dist = distr(weights, gamma, num_actions)
        choice = draw(prob_dist)
        reward = reward_func(choice, t)
        scaled_reward = (reward - reward_min) / (reward_max - reward_min)
        estimated_reward = scaled_reward / prob_dist[choice]
        weights[choice] *= np.exp(gamma * estimated_reward / num_actions)
        yield choice, reward, estimated_reward, weights.copy()
        t += 1

# Simulation Function
def run_simulation(num_actions: int, num_rounds: int, gamma: float) -> np.ndarray:
    """Run a single simulation of the EXP3 algorithm."""
    # Generate biases (mean reward probabilities)
    biases = 1.0 / np.arange(2, num_actions + 2)
    # Generate reward vector using vectorized operations
    reward_vector = (np.random.random((num_rounds, num_actions)) < biases).astype(int)
    reward_func = lambda choice, t: reward_vector[t, choice]

    # Find best action based on total rewards
    best_action = np.argmax(np.sum(reward_vector, axis=0))
    regrets = np.zeros(num_rounds, dtype=float)
    cumulative_reward = 0.0
    best_action_cumulative_reward = 0.0

    for t, (choice, reward, _, _) in enumerate(exp3(num_actions, reward_func, gamma)):
        cumulative_reward += reward
        best_action_cumulative_reward += reward_vector[t, best_action]
        regrets[t] = best_action_cumulative_reward - cumulative_reward
        if t >= num_rounds - 1:
            break

    return regrets
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Log-space EXP3 algorithm (Type 2 from Lattimore)
def exp3_logspace(num_actions: int, reward_func, eta: float, reward_min: float = 0.0, reward_max: float = 1.0):
    """Generator for the EXP3 algorithm using log-space formulation."""
    S = np.zeros(num_actions, dtype=float)
    t = 0
    while True:
        # Compute softmax probabilities from log-weights
        logits = eta * S
        max_logit = np.max(logits)  # For numerical stability
        exp_logits = np.exp(logits - max_logit)
        prob_dist = exp_logits / np.sum(exp_logits)

        # Draw action and get reward
        choice = np.random.choice(num_actions, p=prob_dist)
        reward = reward_func(choice, t)
        scaled_reward = (reward - reward_min) / (reward_max - reward_min)

        # Loss-based estimate (as in Lattimore's log-space version)
        for i in range(num_actions):
            if i == choice:
                S[i] += 1-((1 - scaled_reward) / prob_dist[i])
            else:
                S[i] += 1  # Add default loss of 1

        yield choice, reward, S.copy(), prob_dist.copy()
        t += 1

# Simulation function for log-space EXP3
def run_logspace_simulation(num_actions: int, num_rounds: int, eta: float) -> np.ndarray:
    biases = 1.0 / np.arange(2, num_actions + 2)
    reward_vector = (np.random.random((num_rounds, num_actions)) < biases).astype(int)
    reward_func = lambda choice, t: reward_vector[t, choice]
    
    best_action = np.argmax(np.sum(reward_vector, axis=0))
    regrets = np.zeros(num_rounds, dtype=float)
    cumulative_reward = 0.0
    best_action_cumulative_reward = 0.0

    for t, (choice, reward, _, _) in enumerate(exp3_logspace(num_actions, reward_func, eta)):
        cumulative_reward += reward
        best_action_cumulative_reward += reward_vector[t, best_action]
        regrets[t] = best_action_cumulative_reward - cumulative_reward
        if t >= num_rounds - 1:
            break

    return regrets

# Parameters
num_actions = 10
num_rounds = 10000
num_runs = 1000
eta = 0.07

# Run simulations and average regret
all_regrets_log = np.array([run_logspace_simulation(num_actions, num_rounds, eta) for _ in range(num_runs)])
expected_regret_log = np.mean(all_regrets_log, axis=0)

# Theoretical regret bound
t_values = np.arange(1, num_rounds + 1)
theoretical_bound = (np.e - 1) * np.sqrt(2 * num_actions * t_values * np.log(num_actions))

# Visualization
plt.figure(figsize=(10, 6))
plt.plot(t_values, expected_regret_log, label='Expected Regret (Log-space EXP3)', color='green', linewidth=2)
plt.plot(t_values, theoretical_bound, label='Theoretical Bound', color='red', linestyle='--', linewidth=2)
plt.xlabel('Time Step (t)', fontsize=12)
plt.ylabel('Cumulative Regret', fontsize=12)
plt.title('EXP3 (Log-space) Algorithm: Expected Regret vs Theoretical Bound', fontsize=14)
plt.legend(fontsize=10)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()

# Parameters box
params = f'Parameters:\n- num_actions: {num_actions}\n- num_rounds: {num_rounds}\n- num_runs: {num_runs}\n- eta: {eta:.3f}'
plt.text(0.02, 0.98, params, transform=plt.gca().transAxes, fontsize=10, verticalalignment='top',
         bbox=dict(boxstyle='round,pad=0.5', facecolor='white', edgecolor='gray', alpha=0.9))

# Save plot
plt.show()
plt.savefig('exp3_logspace_regret_plot.png', dpi=300, bbox_inches='tight')
plt.close()

```
